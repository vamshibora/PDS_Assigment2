```python
#(a) Tackled the substitution of absent values in the 'Mileage', 'Engine', 'Power', 'Seats', and 'New_Price' columns. To address these missing values:
#For 'Mileage', 'Engine', and 'Power', we selected the median to minimize the influence of outliers.
#For 'Seats', which has only two unique values (5 or 7), we employed the mode for substitution.
#The 'New_Price' column was eliminated entirely due to its considerable number of missing values.
```

```python
#a)Examining for NULL values within a dataframe
trained_dataset.isna().sum()
```

```python
trained_dataframe = pd.DataFrame(trained_dataset)
```

```python
#b)mileage, engine, power and new price,has units.
trained_dataframe.head()
```

```python
# Extracting numeric values as floats from Mileage
trained_dataframe['Mileage'] = trained_dataframe['Mileage'].str.extract('(\d+\.\d+)').astype(float)

# Removing ' CC' and converting 'Engine' values to floats
trained_dataframe['Engine'] = trained_dataframe['Engine'].str.replace(' CC', '').astype(float)

# Extracting numeric values as floats from 'Power'
trained_dataframe['Power'] = trained_dataframe['Power'].str.extract('(\d+\.\d+)').astype(float)

# Extracting numeric values as floats from 'New_Price'
trained_dataframe['New_Price'] = trained_dataframe['New_Price'].str.extract('(\d+\.\d+)').astype(float)
```

```python
#Substitute the missing values (NA) in the Mileage, Engine, and Power columns with the mean.
trained_dataframe['Mileage'].fillna(trained_dataframe['Mileage'].median(),inplace=True)
trained_dataframe['Engine'].fillna(trained_dataframe['Engine'].median(),inplace=True)
trained_dataframe['Power'].fillna(trained_dataframe['Power'].median(),inplace=True)
trained_dataframe['Seats'].fillna(trained_dataframe['Seats'].mode()[0],inplace=True)
trained_dataframe.drop('Price', axis=1, inplace=True)
```

```python
output_csv_file = '/content/cleaned_dataset.csv'
trained_dataframe.to_csv(output_csv_file, index=False)
```
